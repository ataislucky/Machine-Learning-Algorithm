```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(41)
```

```python
x = 6 * np.random.rand(100, 1) - 3
y = 0.5 * x**2 + x + 2 + np.random.rand(100, 1)
```

```python
poly_feature = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_feature.fit_transform(x)
```

```python
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
```

```python
x_new = np.linspace(-3, 3, 100).reshape(100, 1)
x_new_poly = poly_feature.transform(x_new)
y_new = poly_reg.predict(x_new_poly)
```

```python
plt.scatter(x, y, alpha=0.5, label="Original Data")
plt.plot(x_new, y_new, color="green", linewidth=2, label="Polynomial Regression")
plt.title("Polynomial Regression in Python")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
