```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
exam_scores = np.random.normal(loc=70, scale=10, size=100)
exam_scores
```

```python
plt.hist(exam_scores, bins=30, edgecolor='black')
plt.title("Normal Data Distribution in Python")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
