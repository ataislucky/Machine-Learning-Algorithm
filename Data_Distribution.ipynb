```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.normal(loc=0, scale=1, size=100)
data
```

```python
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histrogram of Data Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
mean_value = np.mean(data)
median_value = np.median(data)
std_deviation = np.std(data)

print(f"Nilai Mean adalah {mean_value}")
print(f"Nilai Median adalah {median_value}")
print(f"Nilai Standar Deviasi adalah {std_deviation}")
```
